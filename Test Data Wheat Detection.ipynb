```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import csv
import cv2

num_rows, num_cols = 10, 1
scale = 16
figsize = (num_rows * scale, num_cols * scale)
_, axes = plt.subplots(num_rows, num_cols, figsize=figsize)
axes = axes.flatten()

fig_i = 0

with open(r'C:\Users\moodl\OneDrive - Singapore University of Technology and Design\Documents\SUTD\Term 7\Computer Vision\CV Project\global-wheat-detection\submission.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        predict = row[1].split()
        row[1] = predict
        if row[0] == "image_id":
            continue
        else:
            #print(row[0])
            test_img = r'C:\Users\moodl\OneDrive - Singapore University of Technology and Design\Documents\SUTD\Term 7\Computer Vision\CV Project\global-wheat-detection\test\{}.jpg'.format(row[0])
            img = mpimg.imread(test_img)
            for i in range(0, len(row[1]), 5):
                xmin = int(row[1][i+1])
                ymin = int(row[1][i+2])
                xmax = xmin+int(row[1][i+3])
                ymax = ymin+int(row[1][i+4])
                score = row[1][i]

                cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (220, 0, 0), 2)
                cv2.putText(img,"score:{}".format(score), (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
            
            cv2.imwrite(r'C:\Users\moodl\OneDrive - Singapore University of Technology and Design\Documents\SUTD\Term 7\Computer Vision\CV Project\global-wheat-detection\{}.jpg'.format(row[0]), img)
            axes[fig_i].imshow(img)
            fig_i += 1
```
